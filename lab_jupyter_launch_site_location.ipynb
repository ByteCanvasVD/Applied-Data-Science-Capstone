```python
import folium
import pandas as pd

# Example launch site data
data = {
    'Site_Name': ['Cape Canaveral', 'Kennedy Space Center', 'Vandenberg Space Force Base'],
    'Latitude': [28.5623, 28.5721, 34.6328],
    'Longitude': [-80.5774, -80.6480, -120.6104]
}

launch_sites = pd.DataFrame(data)

# Calculate the average latitude and longitude for centering the map
average_lat = launch_sites['Latitude'].mean()
average_lon = launch_sites['Longitude'].mean()

# Create a Folium map centered around the average coordinates
launch_map = folium.Map(location=[average_lat, average_lon], zoom_start=5)

# Add markers for each launch site
for idx, site in launch_sites.iterrows():
    folium.Marker(
        location=[site['Latitude'], site['Longitude']],
        popup=site['Site_Name'],
        icon=folium.Icon(color='blue', icon='rocket')
    ).add_to(launch_map)

# Display the map
launch_map


```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import folium
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians

# Function to calculate the distance between two points
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Radius of Earth in km
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Initialize the map centered at NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Add Mouse Position to get coordinates on hover
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Coordinates of the railway point (example)
railway_lat = 34.63494
railway_lon = -120.62429

# Coordinates of the launch site (example for VAFB SLC-4E)
launch_site_lat = 34.632834
launch_site_lon = -120.610746

# Calculate the distance from the launch site to the railway point
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)

# Create and add a marker for the railway point
railway_coordinate = [railway_lat, railway_lon]
folium.Marker(
    location=railway_coordinate,
    popup=f"Railway Point<br>Distance to Launch Site: {round(distance_railway, 2)} km",
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Draw a PolyLine between the launch site and the railway point
folium.PolyLine([railway_coordinate, [launch_site_lat, launch_site_lon]], color='blue').add_to(site_map)

# Display the map
site_map

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
